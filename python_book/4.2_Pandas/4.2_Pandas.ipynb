```python
# 列の場合はaxis=1 行の場合はaxis=0
df_s = df_s.drop("DDD", axis=1)
df_s
```

```python
def judge(num):
    if num > 0:
        return "plus"
    elif num < 0:
        return "minus"
    else:
        return "zero"

df_s.loc[:, "EEE"] = df_s.loc[:, "AAA"].apply(judge)
df_s
```
