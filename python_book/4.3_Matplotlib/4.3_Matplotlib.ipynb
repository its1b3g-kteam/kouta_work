```python
np.random.seed(123)
x = np.random.randn(1000)
fig, ax = plt.subplots()

# counts, edges, patches = ax.hist(x, bins=25)
# n:各ビンの度数　bins:ビンの境界値　patches:描画情報
n, bins, patches = ax.hist(x, bins=25)
print(type(n))
print(n)
print(type(bins))
print(bins)

print(bins[:-1])
print(bins[1:])
x_fit = (bins[:-1] + bins[1:]) / 2
print(x_fit)

y = 1000 * np.diff(bins) * np.exp(- x_fit ** 2 / 2) / np.sqrt(2 * np.pi)
ax.plot(x_fit, y)
plt.show()
```

```python
import pandas as pd
import mpl_finance
import matplotlib.dates as mdates

fig, ax = plt.subplots()
mpl_finance.candlestick2_ohlc(ax,opens=[1,4],highs=[5,5], lows=[1,1],closes=[2,3],width=1,colorup='r', colordown='b')
plt.show()
```

```python
# 日経平均のローソク足を表示してみる
# 株価データ読み込み
df = pd.read_csv("I0000_output.csv")
df.head()
```

```python
# 日付型に変換
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')
df.shape
df
```

```python
df_ = df.copy()
# 日付を数値に変換
df_['date'] = mdates.date2num(df_['date'])
print(df_['date'])
# インデックスを解除してnumpy.ndarrayにする drop=Trueを指定しないとindex(0 1 2 ...)の列が勝手に追加されてしまう
data = df_.reset_index(drop=True).values
print(data)
print(data.shape)
```

```python
fig = plt.figure(figsize=(20, 5))
ax = fig.add_subplot(1, 1, 1)

mpl_finance.candlestick_ohlc(ax, data, width=2, alpha=1, colorup='r', colordown='b')

# ax.grid()

locator = mdates.AutoDateLocator()
ax.xaxis.set_major_locator(locator)
ax.xaxis.set_major_formatter(mdates.AutoDateFormatter(locator))
plt.show()
```

```python
df2 = df.copy()

# df.resample をこのまま実行すると以下のエラーとなる
# Only valid with DatetimeIndex, TimedeltaIndex or PeriodIndex, but got an instance of 'RangeIndex'
# DatetimeIndexに変換してからじゃないとresamapleが使用できなので変換する
dateTimeIndex = pd.DatetimeIndex(df2['date'])
print(type(dateTimeIndex))
df2.index = dateTimeIndex

d_ohlcv = {'open': 'first',
           'high': 'max',
           'low': 'min',
           'close': 'last',
           'volume': 'sum'}
df_w = df2.resample('W-MON', closed='left', label='left').agg(d_ohlcv).copy()
df_w.index = mdates.date2num(df_w.index)
data_w = df_w.reset_index().values

fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(12, 4), sharex=True,
                         gridspec_kw={'height_ratios': [3, 1]})

mpl_finance.candlestick_ohlc(ax[0], data_w, width=4, alpha=0.75, colorup='r', colordown='b')
# 移動平均追加　途中で切れているのはゴールデンウィークで10日間足がないのが原因
ax[0].plot(df_w.index, df_w['close'].rolling(5).mean())
ax[0].plot(df_w.index, df_w['close'].rolling(13).mean())
ax[0].plot(df_w.index, df_w['close'].rolling(26).mean())
# ax.grid()

# 出来高追加
ax[1].bar(df_w.index, df_w['volume'], width=4, color='navy')

locator = mdates.AutoDateLocator()
ax[0].xaxis.set_major_locator(locator)
ax[0].xaxis.set_major_formatter(mdates.AutoDateFormatter(locator))
plt.show()
```
